## **Covariances**

```python
cov_matrix = one_hot_encoded_data_and_label.cov()
# Function to apply formatting
def highlight_max_min_row(row):
    '''
    Highlight the highest and lowest values in a row red.
    '''
    is_max = row == row.max()
    is_min = row == row.min()
    colors = ['color: red' if v else '' for v in (is_max | is_min)]
    return colors

# Display styled DataFrame
cov_matrix.to_csv('./styled_cov_matrix.csv', index=True)
# Apply formatting
cov_matrix.style.apply(highlight_max_min_row, axis=1)
```

```python
# Creating the initial DataFrame
positive_negative_data = {'Positive Cov': [],
        'Negative Cov': []}

df2 = pd.DataFrame(positive_negative_data)
# Traverse the covariance matrix with nested loops
for i in range(len(cov_matrix.index)):
    positive_str = ""
    negative_str = ""
    for j in range(len(cov_matrix.columns)):
        covariance = cov_matrix.iloc[i, j]
        if covariance > 0:
          positive_str += cov_matrix.columns[j] + "\n "
        elif covariance < 0:
          negative_str +=cov_matrix.columns[j] + "\n "
    # New row data
    new_row_data = {'Positive Cov': positive_str, 'Negative Cov': negative_str}

    # Add a new row with index name 'Index4'
    df2.loc[cov_matrix.index[i]] = new_row_data
# Display styled DataFrame
df2.to_csv('./cov_positive_negative.csv', index=True)
df2

```

## **Correlations**

```python
corr_matrix = one_hot_encoded_data_and_label.corr()

# Function to apply formatting
def highlight_max_min_row_corr(row):
    '''
    Highlight the highest and lowest values in a row red.
    '''
    is_max = (row == row.max())
    is_min = (row == row.min())
    colors = ['color: red' if v else '' for v in (is_max | is_min)]
    return colors
# Display styled DataFrame
corr_matrix.to_csv('./styled_corr_matrix.csv', index=True)
# Apply formatting
styled_corr_matrix = corr_matrix.style.apply(highlight_max_min_row_corr, axis=1)

# Display styled DataFrame
styled_corr_matrix
```

```python
# Creating the initial DataFrame
threshold = 0.7
positive_negative_data_corr = {'Strong Positive Corr >'+str(threshold): [],
        'Strong Negative Corr <' +str( -1* threshold): []}

df3 = pd.DataFrame(positive_negative_data_corr)

# Traverse the covariance matrix with nested loops
for i in range(len(corr_matrix.index)):
    positive_str = ""
    negative_str = ""
    for j in range(len(corr_matrix.columns)):
        correlation = corr_matrix.iloc[i, j]
        if correlation > threshold:
          positive_str += corr_matrix.columns[j] + ": "+ str(correlation)+"\n "
        elif correlation < -1*threshold:
          negative_str +=corr_matrix.columns[j] + ": "+ str(correlation)+"\n "
    # New row data
    new_row_data = {'Strong Positive Corr >'+str(threshold): positive_str, 'Strong Negative Corr <'+str(-1* threshold): negative_str}

    # Add a new row with index name 'Index4'
    df3.loc[corr_matrix.index[i]] = new_row_data
# Display styled DataFrame
df3.to_csv('./cov_positive_negative_corr.csv', index=True)
df3
```

```python
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
# Read dataset again
df = pd.read_csv("./Score.csv")
# Perform One Hot encoding
one_hot_encoded_data = pd.get_dummies(df, columns = ['Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour'])
# Perform One Hot encoding
one_hot_encoded_data_and_label = pd.get_dummies(df, columns = ['Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour', 'Credit_Score'])

# Find duplicated rows
duplicated_rows = one_hot_encoded_data [one_hot_encoded_data .duplicated()]
print(duplicated_rows)
# List of columns to be used as features
feature_cols = [col for col in one_hot_encoded_data_and_label.columns if col not in ['Credit_Score_Good', 'Credit_Score_Poor', 'Credit_Score_Standard']]

# Features
X = one_hot_encoded_data_and_label[feature_cols]

# Labels
y = one_hot_encoded_data_and_label[['Credit_Score_Good', 'Credit_Score_Poor', 'Credit_Score_Standard']]
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#### **Create function for linear regression**: This function will take dataset features and labels and it will display confusion matrix and the score of the model

```python
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, precision_score, f1_score, recall_score
import numpy as np
def linear_regression_model(X_train, X_test, y_train, y_test):
  # Regressor model
  linear_model= LinearRegression()
  linear_model.fit(X_train, y_train)
  print("Train Score :" + str(linear_model.score(X_train, y_train)))
  print("Test Score :" + str(linear_model.score(X_test, y_test)))
  # Prediction result
  y_pred_test_linear = linear_model.predict(X_test)     # predicted value of y_test
  y_pred_train_linear = linear_model.predict(X_train)   # predicted value of y_train
  # Find the index of the maximum value in y_pred_test for each row
  max_index_test = np.argmax(y_pred_test_linear, axis=1)

  # Create an array to store the predicted values for y_train
  y_pred_test_adjusted = np.zeros_like(y_pred_test_linear)

  # Set the column corresponding to the maximum value in y_pred_test to 1 for each row
  for i, max_index in enumerate(max_index_test):
      y_pred_test_adjusted[i, max_index] = 1
  # Display the classification report and confusion matrix.
  print("Classification Report:")

  print(classification_report(y_test, y_pred_test_adjusted))

  # Compute confusion matrix
  conf_matrix_linear = confusion_matrix(np.argmax(y_test, axis=1), np.argmax(y_pred_test_adjusted, axis=1))
  return conf_matrix_linear, classification_report(y_test, y_pred_test_adjusted)
```

```python
print("-----------------------------------Linear Regression-----------------------------------------------")
conf_matrix_linear, report_linear = linear_regression_model(X_train, X_test, y_train, y_test)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# Read dataset again
df = pd.read_csv("./Score.csv")
# Perform One Hot encoding
one_hot_encoded_data = pd.get_dummies(df, columns = ['Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour'])

# Find duplicated rows
duplicated_rows = one_hot_encoded_data [one_hot_encoded_data .duplicated()]
print(duplicated_rows)
# List of columns to be used as features
feature_cols = [col for col in one_hot_encoded_data.columns if col not in ['Credit_Score']]

# Features
X = one_hot_encoded_data[feature_cols]

# Labels
y = one_hot_encoded_data[['Credit_Score']]

# Apply label encoder to the 'Credit_Score' column
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=0)
```

### **Models Function**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, precision_score, f1_score, recall_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm

def other_models(X_train, X_test, y_train, y_test, model_name ):
  if model_name == "Logistic Regression":
    model = LogisticRegression(multi_class='multinomial', solver='newton-cg')
  elif model_name == "Decision Tree":
    model = tree.DecisionTreeClassifier()
  elif model_name == "Random Forest":
    model = RandomForestClassifier()
  model.fit(X_train,y_train)
  print("Train Score :" + str(model.score(X_train, y_train)))
  print("Test Score :" + str(model.score(X_test, y_test)))
  y_pred_test_logistic = model.predict(X_test)
  y_pred_train_logistic = model.predict(X_train)

  print("Classification Report:")

  print(classification_report(y_test, y_pred_test_logistic))

  # Compute confusion matrix
  conf_matrix = confusion_matrix(y_test, y_pred_test_logistic)
  return conf_matrix,  classification_report(y_test, y_pred_test_logistic)
```

```python
print("-----------------------------------Logistic Regression-----------------------------------------------")
conf_matrix_logistic, report_logistic = other_models(X_train, X_test, y_train, y_test, "Logistic Regression")
print("-----------------------------------Decision Tree-----------------------------------------------")
conf_matrix_tree, report_tree = other_models(X_train, X_test, y_train, y_test, "Decision Tree")
print("-----------------------------------Random Forest-----------------------------------------------")
conf_matrix_svm, report_svm = other_models(X_train, X_test, y_train, y_test, "Random Forest")

# Display confusion matrices
fig, axs = plt.subplots(2, 4, figsize=(15, 5))

# Confusion matrix for Linear Regression
disp_linear = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_linear, display_labels=label_encoder.classes_)
disp_linear.plot(ax=axs[0,0], values_format='d')
axs[0,0].set_title("Linear Regression")
axs[1,0].text(0.5, -0.4, report_linear, transform=axs[1,0].transAxes, ha='center')
axs[1,0].axis('off')  # Hide axes

# Confusion matrix for Logistic Regression
disp_logistic = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_logistic, display_labels=label_encoder.classes_)
disp_logistic.plot(ax=axs[0,1], values_format='d')
axs[0,1].set_title("Logistic Regression")
axs[1,1].text(0.5, -0.4, report_logistic, transform=axs[1,1].transAxes, ha='center')
axs[1,1].axis('off')  # Hide axes

# Confusion matrix for Decision Tree
disp_tree = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_tree, display_labels=label_encoder.classes_)
disp_tree.plot(ax=axs[0,2], values_format='d')
axs[0,2].set_title("Decision Tree")
axs[1,2].text(0.5, -0.4, report_tree, transform=axs[1,2].transAxes, ha='center')
axs[1,2].axis('off')  # Hide axes

# Confusion matrix for SVM
disp_svm = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_svm, display_labels=label_encoder.classes_)
disp_svm.plot(ax=axs[0,3], values_format='d')
axs[0,3].set_title("Random Forest")
axs[1,3].text(0.5, -0.4, report_svm, transform=axs[1,3].transAxes, ha='center')
axs[1,3].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```

## **PCA (Principal Component Analysis )**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Perform Principal Component Analysis
df = pd.read_csv("./Score.csv")
# Perform One Hot encoding
one_hot_encoded_data = pd.get_dummies(df, columns = ['Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour'])
features = (one_hot_encoded_data.columns).tolist();
features.remove("Credit_Score");

# Separating out the features
x = one_hot_encoded_data.loc[:, features].values

# Separating out the target
y = one_hot_encoded_data.loc[:,['Credit_Score']].values

# Standardizing the features
x = StandardScaler().fit_transform(x)

pca = PCA(n_components=4)

principalComponents = pca.fit_transform(x)
print(pca.explained_variance_ratio_)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4'])
# number of components
n_pcs= pca.components_.shape[0]

# get the index of the most important feature on EACH component
# LIST COMPREHENSION HERE
most_important = [abs(pca.components_[i]).argmax() for i in range(n_pcs)]

initial_feature_names = features
# get the names
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]

# LIST COMPREHENSION HERE AGAIN
dic = {'PC{}'.format(i+1): most_important_names[i] for i in range(n_pcs)}

# build the dataframe
df_importance = pd.DataFrame(dic.items())
print(df_importance)
finalDf = pd.concat([principalDf, one_hot_encoded_data[['Credit_Score']]], axis = 1)
finalDf
```
